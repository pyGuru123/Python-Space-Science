# Astronomy Picture of the Day

Astronomy Picture of the Day is a api provided by NASA and Michigan Technological University. According to the APOD website, "Each day a different image or photograph of our universe is featured, along with a brief explanation written by a professional astronomer."

Importing required Modules

```
import requests
from pprint import pprint
from IPython.display import Image
```

An api key is required to access the APOD api, which you can get from the link below

link = https://api.nasa.gov/

```
key = 'your_api_key_here'
```

API Endpoint and parameters

```
url = 'https://api.nasa.gov/planetary/apod?api_key='
main_url = url + key

# The data for which you want the picture
params = {'date':'2020-12-25'}

print(main_url)
```

Making a request

```
r = requests.get(main_url,params=params)
data = r.json()
data
```

APOD Information

```
date = data['date']
info = data['explanation']
img_url = data['url']

print(date)
pprint(info)
```

Downlading & Displaying Image

```
r = requests.get(img_url)
with open('astro-pic.jpg','wb') as f:
    f.write(r.content)

Image('astro-pic.jpg')
```
