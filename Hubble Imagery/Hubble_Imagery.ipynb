# Hubble Imagery



Using hubble api with python to get latest hubble news, glossary and images. You can read the api documentation from the below link.


### http://hubblesite.org/api/documentation

```python
# pip install requests

import requests
```

### Get Hubble news  

In this part we will learn how to get latest hubble news 

```python
# getting the latest news
# url = 'http://hubblesite.org/api/v3/news'

url = 'http://hubblesite.org/api/v3/news'
r = requests.get(url)
data = r.json()
print(data)
```

```python
news_id = data[0]['news_id']
news_id
```

```python
# getting news of our interest
# url = 'http://hubblesite.org/api/v3/news_release/2020-34'

url = 'http://hubblesite.org/api/v3/news_release/' + news_id
r = requests.get(url)
r.json()
```

### Hubble news Glossary

```python
import requests
from pprint import pprint

def fetch_glossary(url):
    r = requests.get(url)
    data = r.json()
    if len(data) != 0:
        return data['definition']
    else:
        return 'No information on this term'
```

```python
body = 'Jupiter'

url = f'http://hubblesite.org/api/v3/glossary/{body}'
pprint(fetch_glossary(url))
```

```python
body = 'asteroid'

url = f'http://hubblesite.org/api/v3/glossary/{body}'
pprint(fetch_glossary(url))
```

### Get Hubble Images

Using the hubble api to get the latest images

```python
# url= 'http://hubblesite.org/api/v3/images'

url = 'http://hubblesite.org/api/v3/images'
r = requests.get(url)
data = r.json()
data
```

```python
# getting image of our interest
# url= 'http://hubblesite.org/api/v3/image/' + str(id_)

id_ = data[0]['id']
url = 'http://hubblesite.org/api/v3/image/' + str(id_)
r = requests.get(url)
data = r.json()
data
```

```python
# description and image url

desc = data['description']
desc
```

```python
# get filename

image_url = data['image_files'][0]['file_url']
print(image_url)

name = 'image'
img_format = image_url.split('.')[-1]
filename = name + '.' + img_format
filename
```

```python
# making a request for the image and saving it

url =  'http:' + image_url
header = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.122 Safari/537.36'}

r = requests.get(url,headers=header, verify=False)
with open(filename,'wb') as f:
    f.write(r.content)
```
