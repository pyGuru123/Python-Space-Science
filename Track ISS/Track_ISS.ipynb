# What is ISS

The International Space Station is a large spacecraft in orbit around Earth. It serves as a home where crews of astronauts and cosmonauts live. The space station is also a unique science laboratory.  The space station is made of parts that were assembled in space by astronauts. It orbits Earth at an average altitude of approximately 250 miles. It travels at 17,500 mph. This means it orbits Earth every 90 minutes. NASA is using the space station to learn more about living and working in space. 

![ISS in orbit](https://wallpaperaccess.com/full/349100.jpg)

The first piece of the International Space Station was launched in November 1998. Crews aboard the ISS are assisted by mission control centers in Houston and Moscow and a payload control center in Huntsville, Ala.

The space station has the volume of a five-bedroom house or two Boeing 747 jetliners. It is able to support a crew of six people, plus visitors. On Earth, the space station would weigh almost a million pounds. Measured from the edges of its solar arrays, the station covers the area of a football field including the end zones. It includes laboratory modules from the United States, Russia, Japan and Europe.

*   The solar array wingspan (240 feet) is about the same length as the world’s largest passenger aircraft, the Airbus A380.
*   The large modules and other pieces of the station were delivered on 42 assembly flights, 37 on the U.S. space shuttles and five on Russian Proton/Soyuz rockets. The space station is 357 feet end-to-end, one yard shy of the full length of an American football field including the end zones.
*   Eight miles of wire connects the electrical power system aboard the space station.

*   The space station travels an equivalent distance to the Moon and back in about a day.

# Locating ISS

Using the same above API we can get the current coordinates of ISS also

url = http://api.open-notify.org/iss-now.json



Requesting ISS position

````
import requests
from datetime import datetime
url = 'http://api.open-notify.org/iss-now.json'

r = requests.get(url)
data = r.json()

dt = data['timestamp']
lat = data['iss_position']['latitude']
lon = data['iss_position']['longitude']

print('Date and time : ',datetime.fromtimestamp(dt))
print('Latitude :',lat)
print('Longitude :',lon)
````

Plotting the location on a world map

````
import cartopy.crs as ccrs
import matplotlib.pyplot as plt

plt.figure(figsize=(10,8))

ax = plt.axes(projection=ccrs.PlateCarree())
ax.stock_img()


# lon lat
# in google maps : lat lon
plt.scatter(float(lon),float(lat),color='blue', marker='o')

plt.show()
````

# Locating ISS in real time

Wait for few seconds to view the output of this cell, further to get a better animation increase the value of the frame argument at line 31

````
import requests
import matplotlib.pyplot as plt
from matplotlib import animation, rc
import time
import cartopy.crs as ccrs

url = 'http://api.open-notify.org/iss-now.json'


# setting up the background
plt.figure(figsize=(10,8))
ax = plt.axes(projection=ccrs.PlateCarree())
ax.stock_img()

plt.ion()
s = plt.scatter([], [], color='blue', marker='o')

# creating a animation function

def animate(i):
    r = requests.get(url)
    data = r.json()

    dt = data['timestamp']
    lat = data['iss_position']['latitude']
    lon = data['iss_position']['longitude']

    s.set_offsets([float(lon), float(lat)])
    time.sleep(3)

anim = animation.FuncAnimation(plt.gcf(), animate,frames=5)
plt.close()

# Note: below is the part which makes it work on Colab
rc('animation', html='jshtml')
anim
````

```
import requests

	ISS_location = (r.json())

	ISS_data.append([ISS_location['timestamp'],
					ISS_location['iss_position']['latitude'],
					ISS_location['iss_position']['longitude']
					])

	# dump copy to file

	df = pd.DataFrame(ISS_data,columns=['timestamp','latitude','longitude',])
	df.to_csv('ISS_location_'+date+'.csv',index=None)

	if len(ISS_data) > 18000:
		break

	time.sleep(60.0 - ((time.time() - start_time) % 60))
```
